```python
import math
def area_volumen_tetraedro(arista):
  area = arista** 2 * math.sqrt(3)
  volumen = math.sqrt(2) / 12 * arista**2
  return area, volumen
arista = int(input("Ingrese arista: "))
a, v = area_volumen_tetraedro(arista)
print("El area es : ", a)
print("El volumen es : ", v)
```
